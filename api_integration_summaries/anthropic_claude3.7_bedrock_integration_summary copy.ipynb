```python
# Import Libraries
import boto3
import os
import openai
import pandas as pd
import json
import dotenv
```

```python
# Run a Test Prompt with Post Request (Working)
# 1. Initialize Bedrock Runtime client (uses your existing AWS credentials)
bedrock_runtime = boto3.client('bedrock-runtime', region_name='us-east-1')

# 2. Define Claude 3.5 Sonnet model and system prompt
model_id = 'anthropic.claude-3-7-sonnet-20240620-v1:0'

messages = [
    {"role": "user", "content": "Explain the difference between Claude 3 and GPT-4 in two lines."}
]

payload = {
    "anthropic_version": "bedrock-2023-05-31",
    "messages": messages,
    "max_tokens": 400,
    "temperature": 0.7,
    "top_k": 250,
    "top_p": 1
}

# 3. Invoke Claude 3 model via Bedrock
response = bedrock_runtime.invoke_model(
    modelId=model_id,
    contentType='application/json',
    accept='application/json',
    body=json.dumps(payload)
)

# 4. Parse response
response_body = json.loads(response['body'].read())
output_text = response_body['content'][0]['text']

# 5. Display the result
print(output_text)
# This worked successfully.
```
