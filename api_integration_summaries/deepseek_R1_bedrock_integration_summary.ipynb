````python
# Import Libraries
import boto3
import pandas as pd
import json
````

````python
# Run a Test Prompt with Post Request (Working)
# 1. Initialize Bedrock Runtime client (uses your existing AWS credentials)

system_prompt = None

bedrock_runtime = boto3.client('bedrock-runtime', region_name='us-east-1')

# 2. Define DeepSeek R1 model and system prompt
model_id = 'us.deepseek.r1-v1:0'

messages = [
    {"role": "user", "content": "Explain the difference between DeepSeek R1 and GPT-4 in two lines."}
]

if system_prompt:
    messages.append({"role": "system", "content": system_prompt})

payload = {
    "messages": messages,
    "stream": False,
    "max_tokens": 32000
}

# 3. Invoke DeepSeek model via Bedrock
response = bedrock_runtime.invoke_model(
    modelId=model_id,
    contentType='application/json',
    accept='application/json',
    body=json.dumps(payload)
)

# 4. Parse response
response_body = json.loads(response['body'].read())

if response_body.get("choices") and response_body["choices"][0].get("message"):
    output_text = response_body["choices"][0]["message"].get("content", "")

# 5. Display the result
print("response_body", response_body)
# print(output_text)
# This worked successfully.
````
